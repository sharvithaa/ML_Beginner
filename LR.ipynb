```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#import dataset
data=pd.read_csv("https://raw.githubusercontent.com/nunnarilabs/ml/master/logistic_regression/insurance_data.csv")
data.head()
```

```python
data['bought_insurance'].value_counts()
```

```python
x=data.drop('bought_insurance',axis=1)
x
```

```python
y=data['bought_insurance'].to_numpy()
y
```

```python
from sklearn import linear_model
```

```python
model=linear_model.LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.3)
```

```python
model.fit(x_train,y_train)
```

```python
print(model.coef_)
print(model.intercept_)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
print("test scorce",model.score(x_test,y_test))
print('train score',model.score(x_train,y_train))
```
