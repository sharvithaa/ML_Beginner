```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
y=data['Salary'].to_numpy()
y
```

```python
'''80-20 20 la 5% for validation and testing one more split is 60-40 40 la 20-v,20-t'''
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
df=pd.read_csv("https://raw.githubusercontent.com/ezioauditore-tech/AI/main/datasets/Real%20estate.csv")
df
```

```python
x=df.drop('Y house price of unit area',axis=1)
x
```

```python
y=df['Y house price of unit area']
```

```python
polynomial_reg=PolynomialFeatures(degree=2,include_bias=True) #to raise degree to independent variable
```

```python
poly_feat=polynomial_reg.fit_transform(x)
poly_feat.shape
```

```python
x_train,x_test,y_train,y_test=train_test_split(poly_feat,y,test_size=0.3,random_state=101)
```

```python
poly_model=LinearRegression()
poly_model.fit(x_train,y_train)
```

```python
y_pred=poly_model.predict(x_test)
```

```python
a=pd.DataFrame({'y_test':y_test,'y_pred':y_pred})
a
```

```python
poly_me=metrics.mean_absolute_error(y_test,y_pred)
poly_mse=metrics.mean_squared_error(y_test,y_pred)
poly_rise=np.sqrt(poly_mse)
```

```python
error=pd.DataFrame({'poly_me':poly_me,'poly_mse':poly_mse,'poly_rise':poly_rise})
error
```
